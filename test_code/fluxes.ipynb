```python
import lal

def make_list_hv(file):
    list_hv = []
    with open(file, 'r') as f:
        data = []
        for line in f:
            if line[0] == '#':
                pass
            elif line == '\n':
                hv = list(map(lambda l: int(l.split('\t')[6]), data))
                list_hv.append(hv)
                data = []
            else:
                data.append(line[:-1])
        if len(data) > 0:
            hv = list(map(lambda l: int(l.split('\t')[6]), data))
            list_hv.append(hv)
    return list_hv

list_hv = make_list_hv("P21CE2E2.conllu")
print(list_hv)
```

```python
dg = lal.graphs.from_head_vector_to_directed_graph(list_hv[0])
rt = lal.graphs.rooted_tree(dg)
fluxes = lal.linarr.dependency_flux_compute(rt)
nb_fluxes = len(fluxes)
for i in range(nb_fluxes):
    print("Flux", i+1)
    print(" Dependencies:", fluxes[i].get_dependencies())
    print(" Left span:", fluxes[i].get_left_span())
    print(" Right span:", fluxes[i].get_right_span())
    print(" Weight:", fluxes[i].get_weight())
    print(" R/L ratio:", fluxes[i].get_RL_ratio())
    print(" W/S ratio:", fluxes[i].get_WS_ratio())
```
